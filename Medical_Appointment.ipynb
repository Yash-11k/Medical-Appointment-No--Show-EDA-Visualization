```python
df.drop(['PatientId', 'AppointmentID'], axis =1 , inplace= True)
df.columns
```

```python
df.AppointmentDay.unique()
```

```python
df.columns
```

```python
# df['AppointmentDay'] = df['AppointmentDay'].str[:10]
# df['ScheduledDay'] = df['ScheduledDay'].str[:10]





df['ScheduledDay'] = pd.to_datetime(df['ScheduledDay'])
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay'])


print(df[['ScheduledDay', 'AppointmentDay']].dtypes)
df.head()
```

```python
df['due-days'] = df['AppointmentDay'] - df['ScheduledDay']
df['due-days'] = df['due-days'].dt.days

df.tail()
```

```python
df['due-days'].describe()
```

```python
df[df['due-days'] < 0 ]
```

```python
df.drop(df[df['due-days'] < 0 ].index, inplace = True)
df['due-days'].describe()
```

```python
df.columns
```

```python
df['Gender'] = df['Gender'].astype('category')

df['Gender'].dtypes
```

```python
df.columns
```

```python
cols = ['Hipertension', 'Diabetes', 'Alcoholism', 'SMS_received', 'No-show', 'Scholarship']
df[cols] = df[cols].astype('bool')

df[cols].dtypes
```

```python
df.loc[df['No-show']  == 'Yes' , 'No-show'] = 0
df.loc[df['No-show'] == 'No' , 'No-show'] = 1

df['No-show'] = df['No-show'].astype('bool')
df['No-show'].dtypes
```

```python
df[df['Handcap'] > 1 ]
```

```python
# df.loc[df['Handcap'].np.isin([2,3,4]), 'Handcap' ]  = 1


df.loc[df['Handcap'].isin([2, 3, 4]), 'Handcap'] = 1  #remove unnecessary values like 2,3,4 and the 0 = handicap , 1 = no handicap

df['Handcap']  = df ['Handcap'].astype(bool)

df['Handcap'].unique()
```

```python
df.head(3)
```

```python
df[df['Age'] < 0]
```

```python
df.drop(df[df['Age'] < 0 ].index , inplace = True)
df[df['Age'] < 0 ]

df.head()
```
